```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_cidades = pd.read_excel('csv/apuracaoCidade_1_e_2.xls')
df_cidades.head()
```

```python
df_cidades.info()
```

```python
regiao_metropolitana = ['JOAO PESSOA', 'BAYEUX', 
                        'SANTA RITA', 'CABEDELO',
                        'PEDRAS DE FOGO', 'CONDE',
                        'RIO TINTO','CAAPORA',
                        'ALHANDRA', 'PITIMBU',
                        'CRUZ DO ESPIRITO SANTO', 'LUCENA']

df_regiao_metropolitana = df_cidades[(df_cidades['uf'] == 'PB') & 
                                     (df_cidades['nome'].isin(regiao_metropolitana))]

df_regiao_metropolitana.info()
```

```python
abstencao_1 = sns.barplot(x='faltou_1', y='nome', data=df_regiao_metropolitana, orient='h', ascending=False)
plt.title('AbstenÃ§Ã£o na RegiÃ£o Metropolitana de JoÃ£o Pessoa - 1Âº Turno')
```
